**Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Import Dataset**

```python
dataset= pd.read_csv('drinkMenu.csv')
 
```

**Data Preparation**

```python
dataset.drop_duplicates()
```

```python
dataset.fillna(method='backfill')
```

```python
dataset.drop(dataset.columns.difference(['Beverage','Calories',' Sugars(g)']), 1, inplace=True)
```

**Data Visualizations**

```python
dataset = pd.DataFrame(dataset, columns=["Beverage", "Calories"])
 
# plot the dataframe
dataset.plot(x="Beverage", y=["Calories"], kind="bar", figsize=(70, 4.8))
plt.xlabel("Beverage")
plt.ylabel("Calories")
```

```python
sugar = dataset[" Sugars (g)"]
plt.hist(sugar)

 
```
